```python
import pandas as pd
df_100 = pd.read_json('result_mc100.json')
df_100["nb"] = 100
df_500 = pd.read_json('result_mc500.json')
df_500["nb"] = 500
df_1000 = pd.read_json('result_mc1000.json')
df_1000["nb"] = 1000
df_2000 = pd.read_json('result_mc2000.json')
df_2000["nb"] = 2000
df_5000 = pd.read_json('result_mc5000.json')
df_5000["nb"] = 5000
df_10000 = pd.read_json('result_mc10000.json')
df_10000["nb"] = 10000
df_all = pd.concat([df_100, df_500, df_1000, df_2000, df_5000, df_10000])
```

```python
df_all = df_all.loc[df_all['baseline'] != 0]
```

```python
df_all
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#  stack  nb  baseline
grouped = df_all.groupby(['nb', 'baseline'])['stack'].mean().unstack()
grouped_std = df_all.groupby(['nb', 'baseline'])['stack'].std().unstack()

#  grouped bar chart nb  baseline=0~7 
grouped.plot(kind='bar', figsize=(10, 6), yerr=grouped_std)

plt.title("Average Stack by Number of Simulations (nb_simulation) and Baseline")
plt.xlabel("Number of Simulations (nb_simulation)")
plt.ylabel("Average Stack")
plt.legend(title="Baseline", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  loss_history.txt
loss_history = np.loadtxt('loss_history.txt')

# 
plt.figure(figsize=(10, 5))
plt.plot(loss_history, label='Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('DQN Training Loss Curve')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
window = 50
smoothed = np.convolve(loss_history, np.ones(window)/window, mode='valid')
plt.figure(figsize=(10, 5))
plt.plot(smoothed, label=f'Loss (Moving Avg {window})')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('DQN Training Loss Curve (Smoothed)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
